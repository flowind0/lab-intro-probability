**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
import numpy as np
import math

# Given data
seats = 450
tickets_sold = 460
p_show = 0.97

# Normal approximation to Binomial
mean = tickets_sold * p_show
std = np.sqrt(tickets_sold * p_show * (1 - p_show))

# Z-score with continuity correction
z = (seats + 0.5 - mean) / std

# Normal CDF using math.erf (robust)
probability = 0.5 * (1 + math.erf(z / math.sqrt(2)))

print("Probability all passengers get seats:", probability)
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python

# Probability of success on one attempt
p_success = 0.3
p_failure = 1 - p_success

# At least three attempts = first two fail
probability = p_failure ** 2

print("Probability at least three attempts are needed:", probability)
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
# Given data
lambda_hour = 500
capacity = 550

# Normal approximation to Poisson
mean = lambda_hour
std = np.sqrt(lambda_hour)

# Probability server is overwhelmed in one hour
z = (capacity + 0.5 - mean) / std
prob_overload_hour = 1 - (0.5 * (1 + math.erf(z / np.sqrt(2))))

print("Probability of overload in one hour:", prob_overload_hour)

```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
# Probability of overload at least once in 24 hours
prob_overload_day = 1 - (1 - prob_overload_hour) ** 24


print("Probability of overload in 24 hours:", prob_overload_day)
```

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
# Average arrival: one customer every 10 minutes
lambda_rate = 1 / 10  # per minute

# Probability next customer arrives within 5 minutes
prob_within_5 = 1 - np.exp(-lambda_rate * 5)

print("Probability next customer arrives within 5 minutes:", prob_within_5)
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
# Probability no customer arrives for 15 minutes (employee takes break)
prob_break = np.exp(-lambda_rate * 15)

print("Probability employee can take a break:", prob_break)
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
def normal_cdf(z):
    return 0.5 * (1 + math.erf(z / math.sqrt(2)))

mean = 150
std = 10

# Z-scores
z_low = (140 - mean) / std
z_high = (160 - mean) / std

# Normal CDF
cdf_low = 0.5 * (1 + math.erf(z_low / np.sqrt(2)))
cdf_high = 0.5 * (1 + math.erf(z_high / np.sqrt(2)))

probability = cdf_high - cdf_low

print("Probability bird weighs between 140g and 160g:", probability)
```

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
# Mean lifetime = 50 hours
lambda_rate = 1 / 50

# Probability component fails within first 30 hours
probability = 1 - np.exp(-lambda_rate * 30)

print("Probability component fails within 30 hours:", probability)
```
